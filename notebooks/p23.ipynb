```python
nums = [2, 3, 4, 5, 6]

for num in nums:
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num, "is prime")
```

```python
text = input("Enter sentence: ")
words = text.split()

longest = max(words, key=len)
print("Longest word:", longest)
```
