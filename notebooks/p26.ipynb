```python
text = input("Enter sentence: ")
words = text.split()

longest = max(words, key=len)
print("Longest word:", longest)
```

```python
text = input("Enter text: ")
digits = letters = special = 0

for ch in text:
    if ch.isdigit():
        digits += 1
    elif ch.isalpha():
        letters += 1
    else:
        special += 1

print("Digits:", digits)
print("Letters:", letters)
print("Special:", special)
```

```python
a = input("Enter first: ")
b = input("Enter second: ")

if a.lower() == b.lower():
    print("Same")
else:
    print("Different")
```
