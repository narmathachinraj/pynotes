```python
num = int(input("Enter number: "))
print("Binary =", bin(num)[2:])
```

```python
text = input("Enter text: ")
upper = 0
lower = 0

for ch in text:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1

print("Uppercase:", upper)
print("Lowercase:", lower)
```

```python
nums = [1, 2, 3, 4, 5, 6]
total = 0

for i in nums:
    if i % 2 == 0:
        total += i

print("Sum of even numbers:", total)
```
