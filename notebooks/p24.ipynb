```python
start = int(input("Start: "))
end = int(input("End: "))

for i in range(start, end+1):
    if i % 2 == 0:
        print(i)
```

```python
nums = [1, 2, 2, 3, 2, 4]
x = 2

count = 0
for i in nums:
    if i == x:
        count += 1

print("Count:", count)
```
