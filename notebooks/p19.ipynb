```python
nums = [1, 2, 3, 4]

if nums == sorted(nums):
    print("Sorted")
else:
    print("Not sorted")
```

```python
nums = [1, 2, 4, 5]
n = 5

total = n * (n + 1) // 2
missing = total - sum(nums)

print("Missing number:", missing)
```
