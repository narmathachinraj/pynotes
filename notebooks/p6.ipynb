```python
num = int(input("Enter number: "))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
num = int(input("Enter number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed:", rev)
```

```python
num = int(input("Enter number: "))
count = 0

while num > 0:
    count += 1
    num = num // 10

print("Digits:", count)
```
