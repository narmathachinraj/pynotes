```python
num = int(input("Enter number: "))
square = num * num
sum = 0

while square > 0:
    sum += square % 10
    square //= 10

if sum == num:
    print("Neon number")
else:
    print("Not neon")
```

```python
nums = [2, 3, 4, 5, 6, 7]
total = 0

for num in nums:
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            total += num

print("Sum of primes:", total)
```
