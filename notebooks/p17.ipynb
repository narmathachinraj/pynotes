```python
text = input("Enter text: ")
freq = {}

for ch in text:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

```python
text = input("Enter text: ")
result = ""

for ch in text:
    if ch not in result:
        result += ch

print(result)
```

```python
num = int(input("Enter number: "))
sum = 0

for i in range(1, num):
    if num % i == 0:
        sum += i

if sum == num:
    print("Perfect number")
else:
    print("Not perfect")
```
